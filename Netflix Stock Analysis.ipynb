```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df= pd.read_csv('Netflix py.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Date']=pd.to_datetime(df['Date'])
df = df.set_index('Date')
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df,x='Date',y='Volume',label='Volume',color='purple')
plt.title('Volume of Stock vs Time',color='red')
plt.show()
```

```python
df.plot(y=['High','Low','Close'])
plt.title('Netflix Stock Price',color='red')
plt.show()
```

```python
fig, (ax1,ax2,ax3)=plt.subplots(3,figsize=(15,10))
df.groupby(df.index.day).mean().plot(y='Volume',ax=ax1,xlabel='Day',color='red')
df.groupby(df.index.month).mean().plot(y='Volume',ax=ax2,xlabel='Month',color='purple')
df.groupby(df.index.year).mean().plot(y='Volume',ax=ax3,xlabel='Year',color='green')
plt.show()
```

```python
a = df.sort_values(by='High',ascending=False).head(5)
a['High']
```

```python
b = df.sort_values(by='Low',ascending=True).head(5)
b['Low']
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,sharex=True,figsize=(15,6))
fig.suptitle('High & Low Values stock per period of time',color='red')
sns.lineplot(ax=axes[0],y=df['High'],x=df.index,color='green')
sns.lineplot(ax=axes[1],y=df['Low'],x=df.index,color='red')
plt.show()
```
